## Exercise 1 - Perceptron, Adaline and Logistic Regression (10 points)

__NOTE__: Keep your answers short, but provide necessary information.

__Question 1.1 (5 points)__

**Which paramters (input) are needed to compute the weight change of weight ${w}_{j}$ of feature $j$ in case of perceptron, adaline and logistic regression classifier? Mention briefly what these parameters are (do not explain what they do or how they work).**

__Answer 1.1__

_Your answer here:_  
...

<font color='red'>_Solution:_  

The weight change for a single weight is defined as 

$\Delta{w}_{j} = \eta(y^{(i)} - \hat{y}^{(i)})x_{j}^{(i)}$ for perceptron and,


$\Delta{w}_{j} = \eta \sum_{i=1}^n (y^{(i)} - \phi(z^{(i)}))x_{j}^{(i)}$ for adaline and logistic regression.


To compute the weight update we therefore need:

* the learning rate $\eta$
* the target value $y^{(i)}$ of a sample $i$
* $\hat{y}^{(i)}$, $\phi(z^{(i)})$ for sample $i$
* The corresponding value $x_{j}^{(i)}$ of sample $i$ for feature $j$ </font>

__Question 1.2 (2 points)__

**Explain briefly why the equation for the weight update of weight ${w}_{j}$ is slightly different for perceptron compared to those of adaline and logistic regression.**

__Answer 1.2__

_Your answer here:_  
...

<font color='red'>_Solution:_  
The weights of the perceptron are updated after every misclassification during training. The weights of adaline and logistic regression are typically updated only after an epoch or after an an batch-update (i.e. after computing the error across all or a batch of samples, hence $\sum{}$).</font>

__Question 1.3 (3 points)__

**Why does the training of a perceptron sometimes may take longer than the training of an adaline or logistic regression classifier, even though they are all trained on the same data? We assume the same number of epochs for all three classifiers.**

__Answer 1.3__

_Your answer here:_  
...

<font color='red'>_Solution:_  
The perceptron does computations $\Delta{w}_{j} = \eta(y^{(i)} - \hat{y}^{(i)})x_{j}^{(i)}$ for each training sample (one at a time) during one epoch, regardless of misclassification or not (but weights are updated only if $\Delta{w}_{j}$ is not zero). Computations for one training sample at a time is computationally less effective than doing matrix multiplications for the full batch of traing samples as it is done with adaline and logistic regression. With adaline and logistic regression, the weights are typically updated after every epoch, with the gradient descent algorithm trying to minimise the total error across all samples. 
</font>

```python
# Solution to exercise 4.2

class KNN:
    '''
    A class which implements k Nearest Neighbors algorithm from scratch.
    '''
    def __init__(self, k=3, p=2):   
        self.k = k
        self.X_train = None
        self.y_train = None
        self.p = p                
                
    
    def fit(self, X, y):
        '''
        Define training data. No training required with kNN.
        '''
        self.X_train = X
        self.y_train = y
               
    
    def compute_distance(self, r, s): 
        '''
        Compute Minkowski distance.
        '''
        return (np.sum(np.absolute((r - s)) ** self.p)) ** (1/self.p)
    
    
    def predict(self, X):
        '''
        Make predictions for input X data.
        '''
        all_results = []
        for r in X:
            distances = [self.compute_distance(r, s) for s in self.X_train] 
            sorted_k = np.argsort(distances)[:self.k]
            k_nearest = [self.y_train[y] for y in sorted_k]
            all_results.append(stats.mode(k_nearest)[0][0])
            
        return np.array(all_results)
```

__Exercise 5.5.1 (8 points)__

Finally, it is time to tune regression methods and apply them to the test data. Set up a simple grid search (at least three hyperparameter values) for the single parameter indicated for each of the two methods found using the myDataAndMethods()-function above.

```python
# Your code here:
```

```python
# Solution (for candidate numbers giving PLSRegression and SVR):
# 'PCA (n_components) + Linear regression', 'Support Vector regression (C)', 
# 'Ridge regression (alpha)', 'Lasso regression (alpha)',
# 'Elastic Net regression (alpha)', 'Random Forest regression (n_estimators)', 
# 'Decision Tree regressor (max_depth)', 'Partial Least Squares regression (n_components)'
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.cross_decomposition import PLSRegression
from sklearn.svm import SVR

pls_pipe = make_pipeline(PLSRegression(n_components=4, scale=True))
svr_pipe = make_pipeline(StandardScaler(), SVR(gamma=0.1))

param_range1 = list(range(1, X.shape[1]+1))
param_range2 = [0.001, 0.01, 0.1, 0.2, 0.75, 0.9, 1.0]
param_grid1 = [{'plsregression__n_components': param_range1}]
param_grid2 = [{'svr__gamma': param_range2}]

gs1 = GridSearchCV(estimator=pls_pipe, 
                  param_grid=param_grid1, 
                  scoring='r2', 
                  cv=10,
                  n_jobs=-1)
gs1 = gs1.fit(X, y_log)
print(gs1.best_score_)
print(gs1.best_params_)

gs2 = GridSearchCV(estimator=svr_pipe, 
                  param_grid=param_grid2, 
                  scoring='r2', 
                  cv=10,
                  n_jobs=-1)
gs2 = gs2.fit(X, y_log)
print(gs2.best_score_)
print(gs2.best_params_)
```
